```python
import pandas as pd
from sqlalchemy import create_engine

# Konfigurasi koneksi
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "oltp_enggang_khatulistiwa"
DB_USER = "postgres"
DB_PASS = ""

connection_string = f"postgresql+psycopg2://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

try:
    engine = create_engine(connection_string)
    print(f"Engine berhasil dibuat untuk database: {DB_NAME}")
    query = "SELECT * FROM tb_customers LIMIT 10;" 
    df = pd.read_sql_query(query, engine)
    display(df.head())
except Exception as e:
    print("Terjadi kesalahan:", e)
```
